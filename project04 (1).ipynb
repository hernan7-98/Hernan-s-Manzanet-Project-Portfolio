## Task 1: Data distributions

We want to see how the values of data features spead out.

Create the following plot. Your figure should look similar (though not necessarily identical)

```python
columns = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']
df = df_housing[columns]

fig = plt.figure(figsize = (10,7))
ax = fig.add_subplot()

df.hist(ax = ax, color = 'green', alpha = 0.5, bins = 52)
```

## Task 2: Catergory Distribution

We want to catergorize the median income in ranges and compute how many districts for each catergory.

Create the following plot. Your figure should look similar (though not necessarily identical)

![Task 2](./images/task2.png)

```python
df = pd.read_csv("housing.csv")
bins = [0, 1.5, 3, 4.5, 6, float("inf")]
labels = ["(0,1.5]", "(1.5,3]", "(3,4.5]", "(4.5,6]", ">6"]
df["income_range"] = pd.cut(df["median_income"], bins=bins, labels=labels,)
counts = df["income_range"].value_counts()
plt.bar(counts.index, counts.values)
plt.xlabel("Income Range")
plt.ylabel("Number of Districts")
plt.title("Median Income Distribution")
plt.show()
```

## Task 3: Correlation exploration

We want to see the correlations between every two data features

Create the following plot. Your figure should look similar (though not necessarily identical)

![Task 3](./images/task3.png)

```python
ax = sns.pairplot(data = df_housing[['median_house_value','median_income', 'total_rooms', 'housing_median_age']], diag_kind="kde", 
             plot_kws={"alpha": 0.5, "color": "green"},)
```

## Task 4: median_income vs. median_house_value

Looking at the correlation scatterplots, it seems like the most promising attribute to predict the median house value is the median income, so lets zoom in on their scatterplot.

Create the following plot. Your figure should look similar (though not necessarily identical)

![Task 4](./images/task4.png)


```python
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot()
df_housing.plot(kind = 'scatter', ax = ax, x="median_income", y="median_house_value",
           color = 'green', alpha = 0.5, s = 20)
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
```
